```python
#importing the necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#since the data was given in a link we will first put the link in a container called reg_url and reading the csv file using pandas
reg_url = "https://tinyurl.com/54mekfxd"
reg_data = pd.read_csv(reg_url)
print("the  data has been imported successfully")
#printing the first five values in the data
reg_data.head(5)
```

```python
# Plot the data
plt.scatter(reg_data['Hours'], reg_data['Scores'])
plt.xlabel('Hours')  # Set the x-axis label
plt.ylabel('Scores')  # Set the y-axis label
plt.title('Hours vs Scores')  # Set the title of the plot
plt.show()  # Display the plot
```

```python
# Calculate the correlation between 'Hours' and 'Scores'
correlation = reg_data['Hours'].corr(reg_data['Scores'])

# Print the correlation coefficient
print("Correlation coefficient between 'Hours' and 'Scores':", correlation)
```

```python
# Calculate the correlation between 'Hours' and 'Scores'
correlation = reg_data['Hours'].corr(reg_data['Scores'])

# Plot the data points
plt.scatter(reg_data['Hours'], reg_data['Scores'], label='Data Points')

# Add a line of best fit
m, b = np.polyfit(reg_data['Hours'], reg_data['Scores'], 1)  # Calculate the slope and intercept of the line
plt.plot(reg_data['Hours'], m * reg_data['Hours'] + b, color='red', label='Line of Best Fit')

# Set the x-axis and y-axis labels
plt.xlabel('Hours')
plt.ylabel('Scores')

# Set the title of the plot
plt.title('Hours vs Scores')

# Display the correlation coefficient on the plot
plt.text(2, 90, f'Correlation: {correlation:.2f}', fontsize=12)

# Display the legend
plt.legend()

# Show the plot
plt.show()
```

```python
# correlation plot
corr = reg_data.corr()
sns.heatmap(corr, cmap = 'Wistia', annot= True)
```

```python
# Extract the independent variable (X) and dependent variable (y) from the reg_data DataFrame
X = reg_data.iloc[:, :-1].values
y = reg_data.iloc[:, 1].values
```

```python
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)

# Import the LinearRegression model from scikit-learn
from sklearn.linear_model import LinearRegression
```

```python
# Create an instance of the LinearRegression model
lm = LinearRegression()

# Fit the model to the training data
lm.fit(X_train, y_train)

# Print the coefficients of the trained model
print(lm.coef_)
```

```python
# Make predictions on the test set
predictions = lm.predict(X_test)

# Display the predicted values
predictions

# Create a scatter plot to compare the predicted values and the actual values
plt.scatter(y_test, predictions)
plt.ylabel('Predicted')
plt.xlabel('Y test')
```

```python
import sklearn.metrics as metrics
import numpy as np

# Calculate and print the Mean Absolute Error (MAE)
mae = metrics.mean_absolute_error(y_test, predictions)
print('MAE: {}'.format(mae))

# Calculate and print the Mean Squared Error (MSE)
mse = metrics.mean_squared_error(y_test, predictions)
print('MSE: {}'.format(mse))

# Calculate and print the Root Mean Squared Error (RMSE)
rmse = np.sqrt(mse)
print('RMSE: {}'.format(rmse))
```
